```python
# Merge Function Code HOw To mearke array
def merge(arr,left,right):
  i=j=k=0 ## temp array pointer
  while i<len(left) and j<len(right):
    if left[i] < right[j]:
      arr[k]=left[i]
      i+=1
    else:
      arr[k]=right[j]
      j+=1
    k+=1

  while i < len(left):
    arr[k]=left[i]
    i+=1
    k+=1

  while j < len(right):
    arr[k]=right[j]
    j+=1
    k+=1
```

```python
## Merge Sort Algorithm

def merge_sort(arr):
  if (len(arr) > 1):
    mid=len(arr)//2
    left=arr[:mid]## temp array for start to mid
    right=arr[mid:]## temp array for mid to rest
    merge_sort(left)
    merge_sort(right)
    merge(arr,left,right)


```

```python
arr=[38,27,43,3,9,82,10]
print("Given array is :",arr)
merge_sort(arr)
print("Sorted array is:",arr)
```
